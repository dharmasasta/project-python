```python
#8.Find the distribution of the votes and Approx_cost using a user-defined function and for a loop
```

```python
def get_distribution(data, column):
    """
    Returns the distribution of unique values in the given column of the given dataframe.
    """
    distribution = {}
    for value in data[column].unique():
        count = data[data[column] == value][column].count()
        distribution[value] = count
    return distribution
```

```python
votes_distribution = get_distribution(df, "votes")
approx_cost_distribution = get_distribution(df, "approx_cost(for two people)")
```

```python
print("Distribution of votes:")
for value, count in votes_distribution.items():
    print(f"{value}: {count}")
    
print("\nDistribution of approximate cost:")
for value, count in approx_cost_distribution.items():
    print(f"{value}: {count}")
```

```python
#10.Is there any difference b/w the votes of restaurants accepting and not accepting online orders?
```

```python
import scipy.stats as stats

online_votes = df[df["online_order"] == "Yes"]["votes"]
no_online_votes = df[df["online_order"] == "No"]["votes"]

t_stat, p_value = stats.ttest_ind(online_votes, no_online_votes, equal_var=False)

print("t-statistic: ", t_stat)
print("p-value: ", p_value)

if p_value < 0.05:
    print("Reject null hypothesis")
else:
    print("Fail to reject null hypothesis")
```

```python
#12.Find the Best budget Restaurants in any location.
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv("zomoto.csv", encoding="ISO-8859-1")

# Clean the data
df["cuisines"] = df["cuisines"].fillna("Unknown")
df["cuisines"] = df["cuisines"].apply(lambda x: x.lower().strip())

# Extract the cuisines
cuisine_counts = df["cuisines"].str.split(", ", expand=True).stack().value_counts()

# Plot the top 10 cuisines
top_cuisines = cuisine_counts.head(10)
plt.bar(top_cuisines.index, top_cuisines.values)
plt.xticks(rotation=90)
plt.xlabel("Cuisine")
plt.ylabel("Number of Restaurants")
plt.title("Top 10 Cuisines in Bangalore")
plt.show()
```
